-- Applicant's Marital Status.
- Dependents

-- Applicant's eductaion status(If He/She is Graduate/Not-Graduate)
- Self_Employed
-- If the Applicant earns living from their independent pursuit of economic activity.
- ApplicantIncome
Applicant's income earned from employment.

#### Target

- Loan_Status 
-- The loan status indicates where your loan is sanctioned to the Applicant(defined under Yes/No).

```python
loan.isnull().sum()
```

```python
loan.drop(['Loan_ID','Gender','Married','Education'],axis=1,inplace=True)
```

- As above syntax, we Dropped the predictors such as 
- Loan_ID  -- this would make an unwanted changes in the distribution and accuracy in the data.
- Gender and Married -- today every gender and married/unmarried peoples are eligible for the loan and these are not the important factors.
- Education -- Its not true that higher the eductaion higher the income or vice-versa.

```python
loan
```

Using OrdinalEncoder to encode the categorical features into integer type.

```python
from sklearn.preprocessing import OrdinalEncoder
oe= OrdinalEncoder()
```

```python
catcol=loan.select_dtypes(object).columns
loan[catcol]=oe.fit_transform(loan[catcol])
```

```python
loan[catcol]
```

```python
loan.isnull().sum()
```

```python
loan.info()
```

- Filling all the missing values with Median to maximize unwanted distribution

```python
loan['Dependents'].fillna(loan['Dependents'].median(), inplace=True)
loan['Self_Employed'].fillna(loan['Self_Employed'].median(), inplace=True)
loan['LoanAmount'].fillna(loan['LoanAmount'].median(), inplace=True)
loan['Loan_Amount_Term'].fillna(loan['Loan_Amount_Term'].median(), inplace=True)
loan['Credit_History'].fillna(loan['Credit_History'].median(), inplace=True)
loan['Loan_Status'].fillna(loan['Loan_Status'].median(), inplace=True)
```

```python
loan
```

```python
plt.figure(figsize=(10,8))
corr=loan.corr()
sns.heatmap(corr,annot=True,cmap=plt.cm.BuPu)
plt.show()
```

```python
x=loan.iloc[:,:-1]
y=loan.iloc[:,-1]
```

## Train_Test_Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
X_train.shape, X_test.shape
```

## Model Evaluation

```python
# import libraries
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# fit the model
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
#making prediction and calculation of accuracy score
y_pred=model.predict(X_test)
acc_score=accuracy_score(y_test,y_pred)
acc_score
```

So our predictions are over 81% accurate, i.e. we have identified 81% of the loan status correctly.

```python
# importing classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
